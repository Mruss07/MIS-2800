```python
%%sql
DROP TABLE IF EXISTS products;
CREATE TABLE products (
    product_id INTEGER PRIMARY KEY,
    name TEXT,
    price REAL
) STRICT;

INSERT INTO products (name, price) VALUES ('Superbowl Jersey', 120.00);
INSERT INTO products (name, price) VALUES ('Football', 29.99);
INSERT INTO products (name, price) VALUES ('Helmet', 75.00);

SELECT * FROM products;
```

```python
%%sql
SELECT * FROM products WHERE price > 50;
```

```python
%%sql
UPDATE products SET price = 99.99 WHERE name = 'Superbowl Jersey';
DELETE FROM products WHERE product_id = 1;
```
